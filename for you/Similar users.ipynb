## Recommendation Engine

```python
from numpy import dot
from numpy.linalg import norm

def cos_sim(a,b):
    if norm(a)*norm(b) == 0:
        return 'Null'
    return dot(a, b)/(norm(a)*norm(b))
```

```python
users = apps.drop('workshop_suggestions', axis = 1)
users = users.set_index('userid')
```

```python
def find_users(uid, df, similar = 1):
    
    other_users = df.drop(uid, axis = 0)
    other_users['similarity'] = [cos_sim(df.loc[uid], x) for x in other_users.values]
    other_users = other_users[other_users['similarity'] != 'Null']
    if similar == 1:
        other_users = other_users.sort_values('similarity', ascending = False)[0:10]
        return other_users['similarity']
    elif similar == -1:
        other_users = other_users.sort_values('similarity', ascending = True)[0:10]
        return other_users['similarity']
    elif similar == 0:
        other_users = other_users.sort_values('similarity', ascending = True)
        total_users = other_users.shape[0]
        other_users = other_users.iloc[int(total_users/2 - 5):int(total_users/2 + 5)]
        return other_users['similarity']
```

## User search engine

```python
import pickle 

pkl_file = open('encoder.pkl', 'rb')
encoder = pickle.load(pkl_file) 
pkl_file.close()

pkl_file = open('industries.pkl', 'rb')
industries = pickle.load(pkl_file) 
pkl_file.close()

pkl_file = open('technologies.pkl', 'rb')
technologies = pickle.load(pkl_file) 
pkl_file.close()

pkl_file = open('majors.pkl', 'rb')
majors = pickle.load(pkl_file) 
pkl_file.close()
```

```python
majors = [i for i in majors if len(i)>0]
industries = [i for i in industries if len(i)>0]
technologies = [i for i in technologies if len(i)>0]
```

```python
def prep_df(df):
    
    df = df.drop(['other_school', 'workshop_suggestions'], axis = 1)
    df.age_bin = df.age_bin.map({'(15, 18]' : 0, '(18, 20]': 1, '(20, 22]': 2, '(22, 25]': 3, 
                        '(25, 30]' : 4, '(30, 50]': 5})

    df['undergrad'] = df.classification.apply(lambda x: 1 if x in ('Jr', 'Sr', 'So', 'Fr') else 0)
    df['Masters'] = df.classification.apply(lambda x: 1 if x == 'Ma' else 0)
    df['PhD'] = df.classification.apply(lambda x: 1 if x == 'PhD' else 0)
    df.classification = df.classification.map({'O': 0, 'Fr':1 , 'So':2, 'Jr':3, 'Sr':4, 'Ma':5, 'PhD':6})

    df.first_generation = df.first_generation.apply(lambda x: 1 if x==1 else 0)

    df.technology_experience = df.technology_experience.str.replace('[', '')
    df.technology_experience = df.technology_experience.str.replace(']', '')
    df.technology_experience = df.technology_experience.str.replace("'", "")
    df.technology_experience = df.technology_experience.str.replace("'", "")
    df.technology_experience = df.technology_experience.str.replace(" ", "")

    for i in technologies:
        if len(i)>0:
            df[i] = df.technology_experience.apply(lambda x: 1 if i in x else 0)

    df.relavent_industries = df.relavent_industries.str.replace('[', '')
    df.relavent_industries = df.relavent_industries.str.replace(']', '')
    df.relavent_industries = df.relavent_industries.str.replace("'", "")
    df.relavent_industries = df.relavent_industries.str.replace("'", "")
    df.relavent_industries = df.relavent_industries.str.replace(" ", "")

    for i in industries:
        if len(i)>0:
            df[i] = df.relavent_industries.apply(lambda x: 1 if i in x else 0)

    df = df.drop(['technology_experience', 'relavent_industries'], axis = 1)

    df.majors = df.majors.str.replace('[', '')
    df.majors = df.majors.str.replace(']', '')
    df.majors = df.majors.str.replace("'", "")
    df.majors = df.majors.str.replace("'", "")

    for i in majors:
        if len(i)>0:
            df[i] = df.majors.apply(lambda x: 1 if i in x else 0)

    df.minors = df.minors.str.replace('[', '')
    df.minors = df.minors.str.replace(']', '')
    df.minors = df.minors.str.replace("'", "")
    df.minors = df.minors.str.replace("'", "")

    for i in majors:
        if len(i)>0:
            df[i] = df.minors.apply(lambda x: 1 if i in x else 0)

    df = df.drop(['majors','minors'], axis = 1)

    df.num_hackathons_attended = df.num_hackathons_attended.map({'0':0, '1-3':1, '4-7':2, '8-10':3, '10+':4})

    df = pd.concat([df, encoder.transform(df.school)], axis=1, sort=False)
    df = df.drop(['school'], axis =1)
    df = df.set_index('userid')
    return df
```

```python
def find_users2(uid, other_users, similar = 1):
    
    if uid.shape[1]<1:
        print('Choose Any filter')
        return 0
        
    other_users['similarity'] = [cos_sim(uid.values, x) for x in other_users.values]
    other_users = other_users[other_users['similarity'] != 'Null']
    
    if similar == 1:
        other_users = other_users.sort_values('similarity', ascending = False)[0:10]
        return other_users['similarity']
    elif similar == -1:
        other_users = other_users.sort_values('similarity', ascending = True)[0:10]
        return other_users['similarity']
    elif similar == 0:
        other_users = other_users.sort_values('similarity', ascending = True)
        total_users = other_users.shape[0]
        other_users = other_users.iloc[int(total_users/2 - 5):int(total_users/2 + 5)]
        return other_users['similarity']
```

```python
def search_users(df, school = '', majors_vals = '', age_bin = '', classification = '', first_generation = '',
                datascience_experience = '', num_hackathons_attended = '', technology = '', relavent_industries=''):
    
    search = pd.DataFrame(columns = df.columns)
    
    if len(age_bin)>0:
        age = {'(15, 18]' : 0, '(18, 20]': 1, '(20, 22]': 2, '(22, 25]': 3, '(25, 30]' : 4, '(30, 50]': 5}
        search.loc[0, 'age_bin'] = age[age_bin]
    else:
        df = df.drop('age_bin', axis = 1)        
    
    if len(classification)>0:
        search.loc[0, 'undergrad'] = [1 if classification in ('Jr', 'Sr', 'So', 'Fr') else 0]
        search.loc[0, 'Masters'] = [1 if classification == 'Ma' else 0]
        search.loc[0, 'PhD'] = [1 if classification == 'PhD' else 0]
        
        class_dict = {'O': 0, 'Fr':1 , 'So':2, 'Jr':3, 'Sr':4, 'Ma':5, 'PhD':6}
        search.loc[0, 'classification'] = class_dict[classification]
    else:
        df = df.drop(['undergrad', 'Masters', 'PhD', 'classification'], axis = 1)
    
    if len(first_generation)>0:
        search.loc[0, 'first_generation'] = [i if first_generation=='True' else 0]
    else:
        df = df.drop('first_generation', axis = 1)
        
    if len(datascience_experience)>0:
        search.loc[0, 'datascience_experience'] = int(datascience_experience)
    else:
        df = df.drop('datascience_experience', axis = 1)
    
    if len(num_hackathons_attended)>0:
        datahack_dict = {'0':0, '1-3':1, '4-7':2, '8-10':3, '10+':4}
        search.loc[0, 'num_hackathons_attended'] = datahack_dict[num_hackathons_attended]
    else:
        df = df.drop('num_hackathons_attended', axis = 1)
    
    if len(technology)>0:
        for i in technologies:
            search.loc[0, i] = [1 if i in technology else 0][0]
    else:
        df = df.drop(technologies, axis = 1)
    
    if len(majors_vals)>0:
        for i in majors:
            search.loc[0, i] = [1 if i in majors_vals else 0][0]
    else:
        df = df.drop(majors, axis = 1)
    
    if len(school)>0:
        search.loc[0, [i for i in df.columns if 'school' in i]] = \
                    encoder.transform(pd.DataFrame([school], columns = ['school']))
    else:
        df = df.drop([i for i in df.columns if 'school' in i], axis = 1)
    
    if len(relavent_industries)>0:
        for i in industries:
            search.loc[0, i] = [1 if i in relavent_industries else 0][0]
    else:
        df = df.drop(industries, axis = 1)
    
    search = search.dropna(axis = 1)
    
    return find_users2(search, df) 
```
