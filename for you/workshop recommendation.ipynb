```python
def get_workshops_recommendation(diff_0, diff_1, diff_2, DL, DS, Industry, ML, rec_df):
        
    rec_df['rank1'] = 0 
    if diff_0 == 0:
        rec_df['rank1'] = rec_df.difficulty.rank(method = 'dense')
    elif diff_1 == 0:
        rec_df['rank1'] = [1 if i==1 else 4 for i in rec_df.difficulty]
        rec_df.rank1[rec_df.difficulty == 2] = 2
        rec_df.rank1[rec_df.difficulty == 3] = 3
    elif diff_2 == 0:
        rec_df['rank1'] = [1 if i==2 else 4 for i in rec_df.difficulty]
        rec_df.rank1[rec_df.difficulty == 3] = 2
        rec_df.rank1[rec_df.difficulty == 1] = 3
    else:
        rec_df['rank1'] = rec_df.difficulty.rank(method = 'dense', ascending = False)
    
    rec_df['rank2'] = rec_df.track
    rec_df['rank2'] = rec_df['rank2'].map({'DL':DL, 'DS':DS, 'Industry':Industry, 'ML':ML})
    rec_df['rank2'] = rec_df.rank2.rank(method = 'dense', ascending = False)
    
    rec_df = rec_df.sort_values(['rank1', 'rank2']).reset_index(drop = True)
    rec_df['overall_rank'] = rec_df.index + 1
    return rec_df
```

```python
def workshop_recommendations(age_bin, classification, first_generation, datascience_experience, 
                            num_hackathons_attended, technology):
    
    age = {'(15, 18]' : 0, '(18, 20]': 1, '(20, 22]': 2, '(22, 25]': 3, '(25, 30]' : 4, '(30, 50]': 5}
    
    class_dict = {'O': 0, 'Fr':1 , 'So':2, 'Jr':3, 'Sr':4, 'Ma':5, 'PhD':6}
    undergrad = [1 if classification in ('Jr', 'Sr', 'So', 'Fr') else 0][0]
    Masters = [1 if classification == 'Ma' else 0][0]
    PhD = [1 if classification == 'PhD' else 0][0]
    
    f_g = [1 if first_generation=='True' else 0][0]
    
    datahack_dict = {'0':0, '1-3':1, '4-7':2, '8-10':3, '10+':4}

    techs = []
    for i in technologies:
        techs = techs + [1 if i in technology else 0]
        
    feats = [age[age_bin], class_dict[classification], f_g, int(datascience_experience), 
             datahack_dict[num_hackathons_attended], undergrad, Masters, PhD] + techs
     
    feats = np.array(feats).reshape(1, -1)        
    
    diff_0, diff_1, diff_2 = clf1_0.predict(feats), clf1_1.predict(feats), clf1_2.predict(feats)
    
    DL, DS, Industry, ML = clf2.predict_proba(feats)[0,0], clf2.predict_proba(feats)[0,1], \
        clf2.predict_proba(feats)[0,2], clf2.predict_proba(feats)[0,3]
    
    workshops_df = get_workshops_recommendation(diff_0, diff_1, diff_2, DL, DS, Industry, ML, workshops)
    
    return workshops_df
```
